# Live Runtime Monitor

Control and inspect a running NeuroHID service from notebooks.

```python
from neurohid_ml import NeuroHidNotebook

nh = NeuroHidNotebook(
    control_transport="tcp",
    control_host="127.0.0.1",
    control_port=47385,
)
```

```python
connected_stream_id = nh.ensure_connected_stream(rescan=True)
snapshot = nh.snapshot()

{
    "connected_stream_id": connected_stream_id,
    "running": snapshot.get("running"),
    "device_connected": snapshot.get("device_connected"),
    "device_name": snapshot.get("device_name"),
    "signal_quality": snapshot.get("signal_quality"),
    "runtime_mode": snapshot.get("runtime_mode_state"),
    "ml_bridge_connected": snapshot.get("ml_bridge_connected"),
    "ml_bridge_stalled": snapshot.get("ml_bridge_stalled"),
    "decode_latency_p95_us": snapshot.get("decode_latency_p95_us"),
    "action_latency_p95_us": snapshot.get("action_latency_p95_us"),
}
```
