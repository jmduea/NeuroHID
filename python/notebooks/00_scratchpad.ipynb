# NeuroHID Scratchpad

Quick sandbox for ad-hoc experiments.

```python
from neurohid import RuntimeBuilder, SystemConfig
from neurohid_ml import NeuroHidNotebook

config = SystemConfig.from_dict({})
runtime = await RuntimeBuilder(config).start()
nh = NeuroHidNotebook(runtime)
```

```python
connected_stream_id = nh.ensure_connected_stream(rescan=True)
snapshot = nh.snapshot()

{
    "connected_stream_id": connected_stream_id,
    "device_connected": snapshot.get("device_connected"),
    "device_name": snapshot.get("device_name"),
    "signal_quality": snapshot.get("signal_quality"),
    "ml_bridge_connected": snapshot.get("ml_bridge_connected"),
    "ipc_simulated": snapshot.get("ipc_simulated"),
    "runtime_mode_state": snapshot.get("runtime_mode_state"),
    "discovered_streams": snapshot.get("discovered_streams"),
}
```
